# L·ªánh Counter d√πng ƒë·ªÉ ƒë·∫øm s·ªë l∆∞·ª£ng c√°c Emotion

```python
from collections import Counter
counter = Counter()
emotion = all_.values[:,0]
sentences = all_.values[:,1]
counter.update(emotion)
print(counter)
print()
print('Emotion: ',emotion[0:5])
print()
print('Sentences: ',sentences[0:5])
```

# D√πng th∆∞ vi·ªán Underthesea ƒë·ªÉ t√°ch t·ª´ Ti·∫øng Vi·ªát

```python
from underthesea import word_tokenize
print(word_tokenize(sentences[10]))
```

```python
result = [word_tokenize(i) for i in sentences]
print(result)
```

# Fit_on_text cho c√¢u

```python
from keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(result)
print(tokenizer)
```

# ƒê∆∞a c√¢u trong d·ªØ li·ªáu v·ªÅ matrix v·ªõi mode Binary s·ª≠ d·ª•ng Keras

```python
X = tokenizer.texts_to_matrix(result,mode='binary')
print(X)
```

# Chia Train, Test cho d·ªØ li·ªáu

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, emotion, test_size=0.15, random_state=10)
```

```python
X_train
```

# S·ª≠ d·ª•ng Model LogisticRegression c·ªßa Keras ƒë·ªÉ l√†m

```python
from sklearn.linear_model import LogisticRegression  
from sklearn.metrics import classification_report
model_LR = LogisticRegression(max_iter=5000)
model_LR = model_LR.fit(X_train,y_train)
y_pred = model_LR.predict(X_test)
print(classification_report(y_test,y_pred))
```

#  S·ª≠ d·ª•ng Sklearn

```python
all_
```

```python
from pyvi import ViTokenizer
n=[]
for i in all_['Sentence']:
    n.append(ViTokenizer.tokenize(i))
print(n)
```

```python
from sklearn.feature_extraction.text import HashingVectorizer
hv = HashingVectorizer(n_features=16)
features_hv = hv.fit_transform(n).toarray()
print(features_hv)
print(features_hv.shape)
```

```python
y = all_['Emotion'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_hv, y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_model = KNeighborsClassifier(n_neighbors=5)
KNN_model = KNN_model.fit(X_train, y_train)
y_predKNN = KNN_model.predict(X_test)
print(classification_report(y_test,y_predKNN))
```

```python
input_sen = 't·ªôi nghi·ªáp b·∫°n qu√° √† :(('
input_sen = ViTokenizer.tokenize(input_sen)
print(input_sen)
inpu = hv.fit_transform([input_sen]).toarray()
print(inpu)
```

```python
y_pred = KNN_model.predict(inpu)
print(y_pred)
```
