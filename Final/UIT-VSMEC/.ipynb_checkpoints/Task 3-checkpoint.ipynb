# Äá»c 2 file dá»¯ liá»‡u

```python
import pandas as pd
train = pd.read_excel('train_nor_811.xlsx')

valid = pd.read_excel('valid_nor_811.xlsx')

```

# Ná»‘i 2 file excel dÃ¹ng Merge

```python

all_ = pd.merge(train, valid, how='outer')

all_ = all_.drop(['Unnamed: 0'],axis = 1)
```

# Äá»c file dÃ¹ng sá»­a lá»—i chÃ­nh táº£

```python
d = pd.read_excel('replace.xlsx', engine='openpyxl')
d = d.set_index(['word'])["word_replace"].to_dict()
```

```python
d
```

# Thay tháº¿ lá»—i chÃ­nh táº£

```python
all_['Sentence'] = all_['Sentence'].replace(d,regex=True)
```

```python
all_['Sentence'][4092]
```

# Láº¥y dá»¯ liá»‡u

```python
emotion = all_.values[:,0]
sentences = all_.values[:,1]
print()
print(emotion[0:5])
print()
print(sentences[0:5])
```

# TÃ¡ch tá»« Tiáº¿ng Viá»‡t

```python
from underthesea import word_tokenize
result = [word_tokenize(i) for i in sentences]
print(result)
```

```python
from keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(result)
print(tokenizer)
```

# Chuyá»ƒn cÃ¢u vá» ma tráº­n

```python
X = tokenizer.texts_to_matrix(result,mode='binary')
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, emotion, test_size=0.15, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression  
from sklearn.metrics import classification_report
model_LR = LogisticRegression(max_iter=5000)
model_LR = model_LR.fit(X_train,y_train)
y_pred = model_LR.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
comments1 = pd.read_excel('comments_chipu.xlsx',encoding='utf8')

comments2 = pd.read_excel('comments_fobtv.xlsx',encoding='utf8')

comments3 = pd.read_excel('comments_luadao.xlsx',encoding='utf8')

```

```python
accu = 0
for i in range(len(y_predt)):
    if y_predt[i] == all_2.values[:,1][i]:
        accu += 1
print('Accuracy:',accu/len(y_predt)) 
```
